```python
#Masukkan Library
import pandas as pd
import numpy as np
import re
import nltk
import string
import matplotlib.pyplot as plt

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
#Masukkan File
file = 'data.csv'
data = pd.read_csv(file)
data.head()
```

```python
#Text Preprocessing
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# stemming data
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def ScanSentence(text):
    # ubah ke lower case
    text = text.lower()
    # hapus teks tidak relevan
    text = re.sub('[^a-zA-Z]','  ', text)
    # hapus tanda baca
    text = text.translate(str.maketrans("","",string.punctuation))
    # hapus karakter kosong
    text = text.strip()
    # tokenize
    text = word_tokenize(text)
    # hapus stop words
    text = [item for item in text if item not in stop_words]
    # stemming
    text = [stemmer.stem(word=w) for w in text]
    # penggabungan setiap token menjadi kalimat utuh
    text = ' '.join(text)
    return text

data['ScanSentence'] = data['Sentence'].apply(ScanSentence)
data['ScanSentence'].head()
```

```python
#TF-IDF

# import data hasil lemmatization dan stemming, cek data kosong

Sentence = data['ScanSentence']
Sentence.isnull().sum()
```

```python
# import library untuk menghitung kata dengan tf-idf
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
# hitung jumlah kata
cv = CountVectorizer()
term_fit = cv.fit(Sentence)

print (len(term_fit.vocabulary_))
```

```python
# urutkan berdasarkan abjad
term_fit.vocabulary_
```

```python
# operasikan tf-idf
# kolom pertama jumlah dokumen
# kolom kedua letak katanya
# kolom ketiga hasil dari tf-idf

term_frequency_all = term_fit.transform(Sentence)
print (term_frequency_all)
```

```python
# penyelesaian Tf-Idf dengan tfidfVectorizer
# ambil modul TfidfVectorizer dari sklearn
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english') # membuat object
response = vectorizer.fit_transform(Sentence)
print(response)
```

```python
# memunculkan fitur name
vectorizer.get_feature_names_out()
```

```python
# Array 2 dimensi menampilkan hasil Tf-Idf
response.todense()
```

```python
# Hasil Akhir Tf-Idf
import pandas as pd

df = pd.DataFrame(response.todense().T,
                 index = vectorizer.get_feature_names_out(),
                 columns = [f'D{i+1}' for i in range(len(Sentence))])
df
```
